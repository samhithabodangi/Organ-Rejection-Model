## Finding Sequence from Dataset

```python
def find_sequence(allele):
  HLA_allele = allele
  filtered_df = df[df['HLA'] == HLA_allele]

  filtered_rows = df[df['HLA'] == HLA_allele]
  filtered_df["Sequence"]

  return filtered_df
```

```python
def get_sequence(allele):
  if pd.notna(allele):
    s = find_sequence(allele)
    f = pd.Series(s.iloc[0])
  else:
    return None

  return f.loc['Sequence']
```

## Find Mismatches

```python
def find_unique_mismatches(donor_sequence, recipient1_sequence, recipient2_sequence):
    unique_mismatches = []

    for position, donor_aa in enumerate(donor_sequence, start=1):
        recipient1_aa = recipient1_sequence[position - 1]
        recipient2_aa = recipient2_sequence[position - 1]

        if donor_aa != recipient1_aa and donor_aa != recipient2_aa:
            unique_mismatches.append((position, donor_aa, recipient1_aa, recipient2_aa))

    return unique_mismatches
```

```python
def find_mismatches(seq1, seq2):
    """
    Find the mismatches between two sequences of strings.

    Returns:
    - A list of tuples where each tuple contains the index of the mismatch and the corresponding elements.
    """
    mismatches = []
    min_len = min(len(seq1), len(seq2))

    for i in range(min_len):
        if seq1[i] != seq2[i]:
            mismatches.append((i, seq1[i], seq2[i]))

    if len(seq1) != len(seq2):
        for i in range(min_len, max(len(seq1), len(seq2))):
            if len(seq1) > len(seq2):
                mismatches.append((i, seq1[i], None))
            else:
                mismatches.append((i, None, seq2[i]))

    return mismatches

mismatches = find_mismatches(standard, aligned_standard)
print("Mismatches:")
for index, char1, char2 in mismatches:
    print(f"Index {index}: {char1} != {char2}")
```
